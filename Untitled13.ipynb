```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Tharuni's To-Do List</title>
    <!-- Bootstrap CDN for quick styling -->
    <link href="https://maxcdn.bootstrapcdn.com/bootstrap/4.5.2/css/bootstrap.min.css" rel="stylesheet">
    <style>
        body {
            background-color: #f0e68c; /* Light khaki background */
            font-family: 'Arial', sans-serif;
        }
        .todo-container {
            max-width: 500px;
            margin: 50px auto;
            background-color: #ffffff; /* White background for the container */
            padding: 20px;
            border-radius: 10px;
            box-shadow: 0 4px 15px rgba(0, 0, 0, 0.2);
        }
        .todo-input {
            width: 100%;
            padding: 15px;
            border: 2px solid #f39c12; /* Orange border */
            border-radius: 5px;
            margin-bottom: 10px;
            font-size: 16px;
            transition: border-color 0.3s;
        }
        .todo-input:focus {
            border-color: #d35400; /* Darker orange on focus */
            outline: none;
        }
        .todo-list {
            list-style: none;
            padding: 0;
        }
        .todo-item {
            padding: 15px;
            background-color: #ecf0f1; /* Light grey background */
            border-radius: 8px;
            margin-bottom: 8px;
            display: flex;
            justify-content: space-between;
            align-items: center;
            transition: background-color 0.3s;
        }
        .todo-item:hover {
            background-color: #bdc3c7; /* Slightly darker grey on hover */
        }
        .todo-item.completed {
            text-decoration: line-through;
            color: #7f8c8d; /* Grey color for completed tasks */
        }
        .btn-complete {
            background-color: #27ae60; /* Green button */
            color: white;
            border: none;
            margin-right: 10px;
            transition: background-color 0.3s;
        }
        .btn-complete:hover {
            background-color: #219150; /* Darker green on hover */
        }
        .btn-delete {
            background-color: #e74c3c; /* Red button */
            color: white;
            border: none;
            transition: background-color 0.3s;
        }
        .btn-delete:hover {
            background-color: #c0392b; /* Darker red on hover */
        }
    </style>
</head>
<body>

    <div class="todo-container">
        <h3 class="text-center text-warning">âœ¨ Tharuni's To-Do List</h3>
        <input type="text" id="todoInput" class="todo-input" placeholder="Add a new task...">
        <button class="btn btn-info btn-block" onclick="addTask()">Add Task</button>

        <ul id="todoList" class="todo-list mt-4"></ul>
    </div>

    <!-- JavaScript for handling the tasks -->
    <script>
        function addTask() {
            let taskInput = document.getElementById('todoInput');
            let taskText = taskInput.value.trim();

            if (taskText === '') {
                alert('Please enter a task.');
                return;
            }

            let taskList = document.getElementById('todoList');
            let taskItem = document.createElement('li');
            taskItem.className = 'todo-item';

            taskItem.innerHTML = `
                <span>${taskText}</span>
                <div>
                    <button class="btn-complete" onclick="markComplete(this)">Complete</button>
                    <button class="btn-delete" onclick="deleteTask(this)">Delete</button>
                </div>
            `;

            taskList.appendChild(taskItem);
            taskInput.value = ''; // Clear input field after adding
        }

        function markComplete(button) {
            let taskItem = button.parentNode.parentNode;
            taskItem.classList.toggle('completed');
        }

        function deleteTask(button) {
            let taskItem = button.parentNode.parentNode;
            taskItem.remove();
        }
    </script>

</body>
</html>
```
